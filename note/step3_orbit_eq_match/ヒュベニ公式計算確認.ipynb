```python
import pandas as pd
import numpy as np
import math

# [setting parameter]
#==============================================
lat1 = -3.907
lon1 = 141.670
lat2 = -6.70932245254516
lon2 = 142.350814819335

# WGS84 model
a = 6378137                              # the long radius of the Earth
b = 6356752.314245                       # the short radius of the Earth

e = np.sqrt((a*a - b*b)/(a*a)) 
lat1_rad = math.radians(lat1)
lon1_rad = math.radians(lon1)
lat2_rad = math.radians(lat2)
lon2_rad = math.radians(lon2)

dif_lat = lat2_rad - lat1_rad
dif_lon = lon2_rad - lon1_rad

if dif_lon > math.pi:
    dif_lon = 2*math.pi - dif_lon

P = (lat1_rad+lat2_rad) / 2.0                                         # 両点緯度の平均値
W = math.sqrt(1.0-e*e * math.sin(P) * math.sin(P))
M = (a*(1.0 - e*e)) / (W * W * W)
N = a / W

t1 = M * dif_lat
t2 = N * dif_lon * math.cos(P)

dist = math.sqrt(t1 ** 2 + t2 ** 2)
dist = dist / 1000.0                                       # 単位[km] => [m]
print(dist)
#*************************************************
```
